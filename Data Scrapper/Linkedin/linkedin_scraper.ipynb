```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time
import openpyxl
```

```python
# Record the start time
start_time = time.time()

# Create a new Excel workbook
workbook = openpyxl.Workbook()
sheet = workbook.active

# Add headers to the sheet
sheet.append(["User Name", "Job Position", "Company Name", "LinkedIn Profile URL"])
```

```python
# Configure Chrome options
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--disable-notifications")  # Disable notifications

# Use the appropriate path to your ChromeDriver executable
driver = webdriver.Chrome() 

# URL of the LinkedIn post
post_url = input('Enter the LinkedIn Post URL:')
# Open the LinkedIn post URL
driver.get(post_url)

# Click on the "Reactions" link
reactions_link = driver.find_element(By.PARTIAL_LINK_TEXT, "Reactions")
reactions_link.click()

wait = WebDriverWait(driver, 10)

# Wait for the sign-in link to appear
sign_in_link = wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "main__sign-in-link")))
sign_in_link.click()
time.sleep(4)

username_field = driver.find_element(By.ID, "username")
password_field = driver.find_element(By.ID, "password")
```

```python
# Enter your email or phone number into the input field
username_field.send_keys("n7205854@gmail.com")
password_field.send_keys("Burner12345")

sign_in_button = driver.find_element(By.XPATH, "//button[@aria-label='Sign in']")
sign_in_button.click()
time.sleep(10)

reactions_count_link = wait.until(EC.visibility_of_element_located((By.CLASS_NAME, "social-details-social-counts__reactions-count")))
reactions_count_link.click()

# Wait for the dialog box to load
dialog_box = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "artdeco-modal__content")))

scroll_amount = 3000  # Adjust as needed
profiles_per_batch = 50  # Adjust as needed
max_scroll_iterations = 100  # Adjust as needed

loading_indicator_locator = (By.CLASS_NAME, "loading-bar")  # Use the correct class name

last_profiles_count = 0  # Keep track of the last profiles count
processed_urls = set()
# ...

for iteration in range(max_scroll_iterations):
    # Get the current list of profiles
    profiles = dialog_box.find_elements(By.CLASS_NAME, "artdeco-entity-lockup__title")

    # Scroll to the last profile to load more
    last_profile = profiles[-1]
    driver.execute_script("arguments[0].scrollIntoView();", last_profile)

    # Wait for the loading indicator to disappear
    wait.until(EC.invisibility_of_element_located(loading_indicator_locator))

    # Get the updated list of profiles
    updated_profiles = dialog_box.find_elements(By.CLASS_NAME, "artdeco-entity-lockup__title")

    # Check if new profiles were loaded
    if len(updated_profiles) > last_profiles_count:
        new_profiles = updated_profiles[last_profiles_count:]
        for profile in new_profiles:
            print(profile.text)
    
    # Update the last profiles count
    last_profiles_count = len(updated_profiles)

    # Find user elements within the container
    user_elements = dialog_box.find_elements(By.XPATH, "//div[contains(@class, 'social-details-reactors-tab-body')]//div[contains(@class, 'inline-flex full-width')]")

    # Flag to track if any user elements were processed
    user_elements_processed = False

    for user_element in user_elements:
        try:
            user_name_element = user_element.find_element(By.XPATH, ".//span[@dir='ltr']")
            user_name = user_name_element.text
            job_info_element = user_element.find_element(By.XPATH, ".//div[@class='artdeco-entity-lockup__caption ember-view']")
            job_info = job_info_element.text.strip()
            linkedin_profile_link = user_element.find_element(By.TAG_NAME, "a").get_attribute("href")

            # Check if the profile URL has already been processed
            if linkedin_profile_link in processed_urls:
                continue  # Skip processing if the profile is already in the set
            else:
                # Add the profile URL to the set
                processed_urls.add(linkedin_profile_link)

                # Split job_info into job_position and company_name
                if " at " in job_info:
                    job_position, company_name = map(str.strip, job_info.split(" at "))
                elif " @ " in job_info:
                    parts = job_info.split(" @ ")
                    job_position = parts[0].strip()
                    company_name = " @ ".join(parts[1:]).strip()  # Join extra parts into company name
                else:
                    job_position = job_info.strip()
                    company_name = ""

                # Print extracted information
                print("User Name:", user_name)
                print("LinkedIn Profile URL:", linkedin_profile_link)
                print("Job Position:", job_position)
                print("Company Name:", company_name)

                # Append data to the Excel sheet
                sheet.append([user_name, job_position, company_name, linkedin_profile_link])

                # Set the flag to indicate user elements were processed
                user_elements_processed = True

                # Introduce a delay after each iteration
                time.sleep(1)  # Adjust the delay time as needed
        except NoSuchElementException:
            print("User element information not found")

    # Check if no user elements were processed in this batch
    if not user_elements_processed:
        break  # Exit the loop when no user elements were processed

# ...
```

```python
# Save the Excel workbook
excel_file_name = "linkedin_reactions.xlsx"
workbook.save(excel_file_name)

print(f"Data saved to {excel_file_name}")
```
